```python
from IPython.display import clear_output
import time

def player_one_piece():
  while True:
    piece = input("PLAYER 1, SELECT X OR O: ").upper()
    clear_output(wait=True)

    if piece in ["X","O"]:
      return piece
    else:
      print("PLEASE, INSERT X OR O: ")
      clear_output(wait=True)

def player_two_piece(p1p):
  if p1p == 'X':
    return "O"
  else:
    return "X"

def board_printing(board):
    print("      |        |      ")
    print(f"   {board[0]}  |    {board[1]}   |   {board[2]} ")
    print("______|________|______")
    print("      |        |      ")
    print(f"   {board[3]}  |    {board[4]}   |   {board[5]} ")
    print("______|________|______")
    print("      |        |      ")
    print(f"   {board[6]}  |    {board[7]}   |   {board[8]} ")
    print("      |        |      ")

def moves(player_one):
  if player_one:
      player = "ONE"
  else:
      player = "TWO"

  print(f"PLAYER {player}, WHERE DO YOU WANT TO PLACE YOUR PIECE?")

  board_printing(board_example)
  time.sleep(4)
  clear_output(wait=True)

  board_printing(board)
  move = int(input("ENTER YOUR MOVE: "))

  while move not in board_example or move == 'O' or move == "X":
      move = int(input("PLEASE, ENTER A VALID INPUT: "))

  clear_output(wait=True)

  return move - 1

def moves_on_the_board(move, player_one):
  if player_one == True:
    board[move] = p1p
    board_example[move] = p1p
  else:
    board[move] = p2p
    board_example[move] = p2p

def win_check(board, p1p, p2p):
    winning_combinations = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],
        [0, 3, 6], [1, 4, 7], [2, 5, 8],
        [0, 4, 8], [2, 4, 6]
    ]

    for combo in winning_combinations:
        if all(board[i] == p1p for i in combo):
            print("CONGRATULATIONS PLAYER ONE, YOU WON!")
            return True
        elif all(board[i] == p2p for i in combo):
            print("CONGRATULATIONS PLAYER TWO, YOU WON!")
            return True

    if all(i != " " for i in board):
        print("IT'S A DRAW!")
        return True

    return False

def ask_want_to_play():
  while True:
    answer = input("DO YOU WANT TO KEEP PLAYING? (Y OR N): ").upper()
    clear_output(wait=True)

    if answer in ["Y", "N"]:
      return answer
    else:
      print("I DON'T UNDERSTAND, TYPE Y OR N, PLEASE. ")
      clear_output(wait=True)


print("WELCOME TO TIC TAC TOE!")
time.sleep(3)
clear_output(wait=True)

game_on = 'Y'
while game_on == 'Y':
  p1p = player_one_piece()
  p2p = player_two_piece(p1p)

  board_example = [i for i in range(1, 10)]
  board = [" " for i in range(1, 10)]

  player_one = True
  while True:
    move = moves(player_one)
    moves_on_the_board(move, player_one)
    if win_check(board, p1p, p2p):
      game_on = ask_want_to_play()
      break
    if player_one == True:
      player_one = False
    else:
      player_one = True
```
